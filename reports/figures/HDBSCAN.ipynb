```python
#HDBSCAN

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as data

sns.set_context('poster')
sns.set_style('white')
sns.set_color_codes()
plot_kwds={'alpha' : 0.5, 'sss' : 80, 'linewidths' :0}

filepath ='/Users/annabearden/Downloads/Data III/RawData.csv' 
data=pd.read_csv(filepath)
df = pd.DataFrame(data)
import pandas as pd
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA as PCA
from sklearn.datasets import load_digits
digits = load_digits()
test_data=np.vstack(digits)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as data
import hdbscan 

clusterer= hdbscan.HDBSCAN(min_cluster_size=10, gen_min_span_tree=True)
clusterer.fit(df)
```

```python
hdbscan.HDBSCAN(algorithm='best', alpha=1.0, approx_min_span_tree=True,
       gen_min_span_tree=True, leaf_size=40,
       metric='euclifdean', min_cluster_size=5, min_samples=None, p=None)
```

```python
clusterer.minimum_spanning_tree_.plot(edge_cmap='viridis',
                                      edge_alpha=0.6,
                                      node_size=80,
                                      edge_linewidth=2)
```

```python
clusterer.single_linkage_tree_.plot(cmap='viridis', colorbar=True)
```

```python
clusterer.condensed_tree_.plot()
```

```python
clusterer.condensed_tree_.plot(select_clusters=True, selection_palette=sns.color_palette())
```

```python
palette = sns.color_palette()
cluster_colors = [sns.desaturate(palette[col], sat)
                  if col >= 0 else (0.5, 0.5, 0.5) for col, sat in
                  zip(clusterer.labels_, clusterer.probabilities_)]
plt.scatter(*test_data.T, sss=50, linewidths=0, c=cluster_colors, **plot_kwds)
```
